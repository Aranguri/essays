## Introduction
### Measuring
If we don't measure the state of a photon/electron, then it behaves like a wave. Instead, if we measure it, it behaves like a particle. If we have any way of knowing the state of an electron, then it will collapse to some state and behave like a particle. 

Before the measurement, the electron wasn't in any particular state. When we measure, it gets projected and we lost all the previous information for that electron.

We can take atoms and use their ground and first state as representing 0 and 1. While we don't measure the electron, it stays having a complex amplitude, which is a superposition of all states. If we perform operations on the electrons without measuring them, we are working on this amplitudes. {todo: is this real? give example of operations} It's only when we measure it that we get a specific state. The amplitudes tell us the probability for each given state.

We don't need to limit ourselves to a qubit of two states. We can take advantage of all the states of an electron.

### Quantum bit
We have this quantum bit $a|0> + b|0>.$ We can interpret it as a vector $[a, b]$ in a 2D space. $|0>$ is the x-axis, and $|1>$ is the y-axis. Then, a given electron corresponds to a vector $v$ somewhere between the two axis $|0>$ and $|1>.$ We can express the position of the electron in terms of $\theta,$ the angle between the vector and the x-axis. Namely, $v = [cos(\theta), sin(\theta)] = [cos(\theta), cos(\pi/2 - \theta)],$ and in bra-ket notation $v = cos(\theta)|0> + cos(\pi/2 - \theta)|1>.$ Thus, when we measure the electron, it goes to the the axis $|0>$ with probability $cos^2(\theta)$ and it goes to the axis $|1>$ with probability $cos^2(\pi/2 - \theta).$
    
We can also take another basis, and perform a measurement on that basis. So, instead of measuring with |0> and |1>, we measure with |u> and |$u^{\perp}$> where u = $\frac{1}{\sqrt2}$|0> + $\frac{1}{\sqrt2}$|1>. Now, instead of seeing whether the state is in the |0> or |1> position, we see whether it is in the |u> or |$u^{\perp}$> position.

Sign basis (as opposed to the bit basis with |0> and |1>)
$$
|+> := \frac{1}{\sqrt2}|0> + \frac{1}{\sqrt2}|1> \\
|-> := \frac{1}{\sqrt2}|0> - \frac{1}{\sqrt2}|1>
$$

### Two qubits
We can entangle two qubits. In that case, we have an amplitude (complex number) for each possible configuration of the qubits. Now, when we measure one of the two qubits, the probability of being a zero is the sum of the probabilities of all the whole system being 00 and being 01. And after that measurement, we remove the states that the electron can't be in (10 and 11,) and we normalize the states 00 and 01 to sum to probability 1.

There are some configurations of two qubits where we can't decompose it into separate qubits.

Also, if two qubits are close together and interacted, then when we distant them, they continue being entangled.

They don't need to be communicating. When they interacted, they could have decide what bit they were going to say.

### EPR paradox
Local physics: things only interact with the other things that are near

Realism: if we are not measuring/seeing something, it's still there

If we combine both local physics and realism, we get that quantum mechanics is incomplete. The argument is as follows. Take two qubits and separate them by a huge distance. Measuring the first qubit can't affect measuring the second qubit because of the long distance. Then, the two qubits must had decided both the sign and the bit when they interacted with each other. But now, they both already know the sign and the bit. But quantum mechanics tells us that this is impossible. So, ERP paradox concludes that QM is incomplete.

### Rotational invariance
$$
|\phi> = \frac{1}{\sqrt2}(|00> + |11>) \\
= \frac{1}{\sqrt2}(|uu> + |u^\perp u^\perp>)  \\
$$
$$
\ket{u} = a\ket0 + b\ket1 \\
\ket{u^\perp} = - b\ket0 + a\ket1 \\
\ket{uu} + \ket{u^\perp u^\perp} = (a^2 + b^2)\ket{00} + (a^2 + b^2)\ket{11} \\
= \ket{00} + \ket{11} \\
$$

For any unit vector $u$. That is, $\ket\phi$ is invariant under rotation.

Also, when we measure $\ket\phi = \frac{1}{\sqrt2}(\ket{u^\perp u^\perp}),$ we will define both the first and second qubit. If the first qubit is $\ket u$, then the second qubit is $\ket u.$ (The same with $\ket u^\perp.$) Now, we can measure the second qubit in the $v, v^\perp$ basis. Say $u$ and $v$ form an angle of $\theta.$ Then, the probability of $v$ in the second qubit given that we got $u$ in the first qubit is $cos^2(\theta).$ Now, the probability of $v^\perp$ in the second qubit given that we got $u^\perp$ in the first qubit is also $cos^2(\theta).$

### CHSH Game
QM: we can't communicate information between the two entangled qubits, but we can create non-local correlations of ranodm numbers we generate.

Two person that share a qubit. Alice has input x and output a, Bob has input y and output b. They want to coordinate such that $xy = a \oplus b.$ If they measure the qubit using a certain basis that depends on the input, then they can use to their advantage the non-local correlations.

Depending on the value of the qubit, Alice measures it with some basis
* If x=0, she measures in the x, y basis
* If x=1, she measures in the +, - basis
Bob does the same
* If y=0, he measures in one orthogonal basis where the axes are moved to the left by $\pi / 8$
* If y=1, he measures in one orthogonal basis where the axes are moved to the right by $\pi / 8$

The probability of getting the same value after measuring the qubit is as follows.
* x=0 and y=0. Distance: $\pi / 8.$ $P(same) = cos^2(\pi / 8)$
* x=0 and y=1. Distance: $\pi / 8.$ $P(same) = cos^2(\pi / 8)$
* x=1 and y=0. Distance: $\pi / 8.$ $P(same) = cos^2(\pi / 8)$
* x=1 and y=1. Distance: $(3 \pi) / 8.$ $P(diff) = cos^2(\pi / 8)$

## Quantum circuits
The state evolves in time by a rotation in the hilbert space {todo: define hilbert space.} Those rotations can be expressed by multiplying by a unitary matrix (orthogonal?)

The unitary matrix tells us that when we perform the inner product between the same basis vector, we get one. And when we perform the inner product between two different basis vector, we get zero. Thus, we have a normalized and orthogonal basis.

### Gates
Gates are just multiplying a vector by a unitary matrix.

Phase flip (Z): $a\ket0 + b\ket1 = a\ket0 - b\ket1$
$$
[1, 0] \\
[0, -1]
$$

Bit flip (X): $a\ket0 + b\ket1 = b\ket0 + a\ket1$
$$

We can't know both the position and the wavelenth of the particle at a given time. However, we do know a PDF that tells us the probability of a particle being in a given position and time.
