## The real and complex number systems
Between two rational numbers there is always another rational number. Eg $s < (s+r) / 2 < r.$

However, there are some numbers that can't be expresses with rational numbers. Eg 
$$
2 = p^2 \\
2 = \left ( \frac{m}{n}\right )^2 \\
2n^2 = m^2 \\
$$

$$
A = \{p | p^2 < 2 \land p \in Q\} \\
B = \{p | p^2 > 2 \land p \in Q\} \\
$$

Take any p from A. Define q as follows.

$$
q := p - \frac{p^2 - 2}{p + 2} = \frac{2p + 2}{p + 2}\ (1)\\
q^2 - 2 = \frac{2(p^2 - 2)}{(p + 2)^2}\ (2)\\
$$

If $p \in A,$ 

$$
p^2 < 2 \\
-\frac{p^2 - 2}{p + 2} > 0 \\
q > p \\
\frac{2(p^2 - 2)}{(p + 2)^2} < 0\\
q^2 - 2 < 0
q^2 < 2 \\
p^2 < q^2 < 2 \\
$$

If $p \in B,$ then $2 < q^2 < p^2.$ (Analogous proof to that of the set A.)

Say we have a set S of all the rational numbers smaller than 2. Now, there isn't going to be a maximum for this set, because I can take any given maximum m and now I take the average between m and 2, and that is greater than m but it's still in s. We also can take all the rational numbers which their square is smaller than 2, and the same reasoning applies: there is no maximum number in the set. 

### Bounds
S is a set. $E \subset S.$ 

We say that E is bounded above and $a$ is the upper bound of $E$ iff $a \in S$ and $\forall{p \in E}: a \geq p.$

We say that E is bounded below and $a$ is the lower bound of $E$ iff $a \in S$ and $\forall{p \in E}: a \leq p.$

An upper bound u of a set E is the supremum of E if $\forall x < u:$ x isn't an upper bound of E. 

A  lower bound u of a set E is the infimum  of E if $\forall x > u:$ x isn't a  lower bound of E. 

An ordered set S has the least-upper-bound property if it fulfills that

if $E \subset S,$ E isn't empty, and E is bounded above, then sup E is in S.

(Similarly defined for the greatest-lower-bound)

### 
Theorem: If S has the LUB property, then it has the GLB property. 
Proof idea: we want for each set R (which fulfills some properties) to have a greatest lower bound. Take any set T that contains all the numbers that are greater than or equal to every number in R. We have something like this
$$[-|-R-|-T-]$$

By definition, R has a LUB. Say it's $\alpha$
$$[-|-R-\alpha-T-]$$
where $S = -R-\alpha$ and $T = \alpha-T-.$

Now, that LUB will be the GLB of T. 

Proof:

Take any set B that fulfills
* B is not empty
* B is bounded below
* $B \subset S$

$L := \{l | l is a LB of B\}$

As B is bounded below, then L is not empty.

As every element in L is a LB of B, then $L \subset S.$

As every element in B is a UB of L, then L is bounded above.

As S has the LUB property, we can define $\alpha := sup L.$ By definition, $\alpha \in S.$

$\forall y < \alpha:$ y isn't an UB of L. (Because $y < \alpha$ and $\alpha$ is the LUB of L.) Hence no $y < \alpha$ is in B.

Thus, everything in B is greater than or equal to $\alpha.$ Thus, $\alpha$ is a LB of B. Thus, $\alpha \in L.$

L is the set of all lower bounds of B. We know that $\alpha$ is the greatest lower bound of B because it's an upper bound of L. Thus, $\alpha$ is the lower upper bound of B. Alternatively, $\alpha = inf B.$

## Field''s
We say that a set is a field if it has two operations (addition and multiplication) that fulfill the following axioms.

Addition axioms
- If x is in the field, and y is in the field, then x + y is in the field.
- Commutative: x + y = y + x
- Associative: x + (y + z) = (x + y) + z
- Existence of zero: x = x + 0
- Esistence of inverse: For every x, there exists a -x such that x + (-x) = 0

Addition properties
* Cancellation law: x + y = x + z \implies y = z {why not the other way around too?}
* Uniqueness of identity: x + y = x \implies y = 0
* Uniqueness of inverse: x + y = 0 \implies y = -x
* Double negation: -(-x) = x

Proof of properties:
$y = 0 + y = (-x + x) + y = -x + (x + y)$
(cancellation) $= -x + (x + z) = (-x + x) + z = 0 + z = z$
(uniqueness of identity) $= -x + x = 0$
(uniqueness of inverse) $= -x + 0 = -x$

We showed the uniqueness of zero because whenever we have a structure like $x + something = x$ then that something is forced to be 0, and it can't be any other element. (Similarly for the inverse: $x + something = 0$ forces something to be and only to be -x.)

For double negation, we know that $x + y = 0 \implies y = -x.$ Chaning the name of the variables we get $y + x = 0 \implies x = -y.$ Take $y = -x.$ We then have $-x + x = 0 \implies x = -(-x).$ As the first part of the implication is true, the second part is true

Multiplication axioms
- If x, y are in the field, then xy is in the field
- Commutative: xy = yx
- Associative: x(yz) = (xy)z
- Existence of identity: if x\neq0 x = x * 1
- Esistence of inverse: For every $x \neq 0$, there exists a 1/x such that x * 1/x = 1

Distributive axiom: $a(b + c) = ab + ac$

### Properties of a field
Thm: 0x = 0
Proof:
* 0x + 0x = (0 + 0)x = 0x
* 0x = 0 (because of the uniqueness of the identity)

Them: $x \neq 0 \land y \neq 0 \implies xy \neq 0.$
Proof:
* Assume xy = 0
* xy = 0x = x0
* y = 0 (absurd)

Theorem: $(-x)y = -(xy)$
Proof: 
* (-x)y + xy = (-x + x)y = 0y = 0
* -(xy) + xy = 0
* => (-x)y = -(xy)

Theorem: $x(-y) = -(xy)$
Proof:
* x(-y) + xy = x(-y + y) = x0 = 0
* -(xy) + xy = 0
* => x(-y) = -(xy)

### Ordered field
A field F is ordered if it fulfills the two following properties
1) $x, y, z \in F \land y < z \implies x + y < x + z$
2) $x, y \in F \land x > 0 \land y > 0 \implies xy > 0$

#### Properties of an ordered field
Theorem: $x > 0 \iff -x < 0$
Proof idea: we take the first property from an ordered field. Imagine the number line with two dots for y and z fulfilling 0 < y < z. Now, add two dots more to the right of the previous to dots. These dots are x + y and x + z (again they fulfill 0 < x + y < x + z.) Imagine the distance between xy and xz and between y and z as something small, so that you have two groups of two points. Now, y and z are whatever we want. So we move the right tip of the segment y, z to touch the left tip of the segment x + y, x + z. So far we have z = x + y. Notice the segments have length x now (first segment has length x + y - y. second: x + z - z.) Now translate both segments so as to have its center (where the tips touch) at 0. We now have a tip in -x, two tips in 0, and one tip in x. The segments meant that the right tip was greater than the left tip. Note we only assumed x > 0. And now we know -x < 0 (because of the segment to the left of 0.) 
Proof: 
* =>
* $y < z \implies x + y < x + z.$ 
* $y := -x, z := 0 \implies 0 < x$
* <=
* $y < z \implies -x + y < -x + z.$ 
* $y := 0, z := x \implies -x < 0$

Theorem: $x > 0 \land y < z \implies xy < xz$
Proof idea: {add the intuition of multiplying numbers and starting from x, y}
Proof: 
$$
z > y \\
z - y > y - y \\
z - y > 0 \\
x(z - y) > 0 \\
xz - xy > 0 \\
xz > xy \\
$$

Theorem: $x < 0 \land y < z \implies xy > xz$
Proof: 
$$
z > y \\
z - y > y - y \\
z - y > 0 \\
(-x)(z - y) > 0 \\
-(x(z - y)) > 0 \\
x(z - y) < 0 \\
xz - xy < 0 \\
xz < xy \\
$$

In math we generally have an operation and its inverse. A function f has an inverse $f^{-1}$ if $f^{-1}$ fulfills that 
$$
f^{-1}(f(x)) = f(f^-1(x)) = x \\
$$

Let's consider functions that have two variables as input. We are going to define the inverse as the function that returns one of the two input variables. Thus, we have two inverse functions
$$
f^{-1}_{first}(f(x, y), y) = x \\
f^{-1}_{second}(f(x, y), x) = y \\
$$
where the subscript of the function tells us whether we extract the first or the second input variables.

For addition, we have that 
$$
f(x, y) := x + y \\
f^{-1}_{first}(x, y) := x - y\\
f^{-1}_{second}(x, y) := x - y\\
f^{-1}_{first}(f(x, y), y) = f^{-1}_{first}(x + y, y) = (x + y) - y = x \\
f^{-1}_{second}(f(x, y), x) = f^{-1}_{first}(x + y, x) = (x + y) - x = y \\
$$

For multiplication, we have that 
$$
f(x, y) := x * y \\
f^{-1}_{first}(x, y) := x / y\\
f^{-1}_{second}(x, y) := x / y\\
f^{-1}_{first}(f(x, y), y) = f^{-1}_{first}(x * y, y) = (x * y) / y = x \\
f^{-1}_{second}(f(x, y), x) = f^{-1}_{second}(x * y, x) = (x * y) / x = y \\
$$

Addition and multiplication are commutative. Formally,
$$
f(x, y) = f(y, x) \\
f^{-1}_{first}(f(x, y), y) = f^{-1}_{second}(f(y, x), y) \\
\therefore f^{-1}_{first} = f^{-1}_{second} \\
$$

Our next operation, exponentiation, isn't commutative. That is, we have that $f(x, y) \neq f(y, x),$ but we want that $f^{-1}_{first}(f(x, y), y) = f^{-1}_{second}(f(y, x), y).$ Thus, $f^{-1}_{first} \neq f^{-1}_{second}.$

$$
f(x, y) := x ^ y \\
f^{-1}_{first}(x, y) := \sqrt[y]{x} \\
f^{-1}_{first}(f(x, y), y) = f^{-1}_{first}(x ^ y, y) = \sqrt[y]{x ^ y} = x \\
f^{-1}_{second}(x, y) := \log_y(x) \\
f^{-1}_{second}(f(x, y), y) = f^{-1}_{second}(x ^ y, x) = \log_x{x ^ y} = y \\
$$

We successfully extracted the first input variable with $f^{-1}_{first}$ and the second input variable with $f^{-1}_{second}.$
