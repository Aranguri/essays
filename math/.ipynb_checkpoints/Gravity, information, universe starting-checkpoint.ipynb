How information can appear from nothing?

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import itertools
import time
```

```python
def sign(v):
    a = 1 if v[0] > 0 else -1
    b = 1 if v[1] > 0 else -1
    return [a, b]
```

```python
n = 10
steps = 1
alpha = .1
vs = np.concatenate((
    np.random.choice((-1, 1), size=(n, 1)),
    np.random.randn(n, 2)
), 1)

deltas = {i: 0 for i in range(n)}
fig = plt.figure()

for i in itertools.count():
    if i % steps == 0:
        vs1 = [v[1:] for v in vs if v[0] == 1]
        vs2 = [v[1:] for v in vs if v[0] == -1]
        plt.cla()
        plt.xlim(-5, 5)
        plt.ylim(-5, 5)
        if len(vs1): plt.scatter(*zip(*vs1), color='red')
        if len(vs2): plt.scatter(*zip(*vs2), color='blue')
        fig.canvas.draw()
        time.sleep(.001)
    
    for j, v in enumerate(vs):
        for k, u in enumerate(vs):
            if j != k:
                deltas[j] += (1 * v[0] * u[0]) / ((u[1:] - v[1:]) + sign(u[1:] - v[1:]))
                
    for j, v in enumerate(vs):
        v[1] = 5 if v[1] < -5 else v[1]
        v[1] = -5 if v[1] > 5 else v[1]
        
        v[2] = 5 if v[2] < -5 else v[2]
        v[2] = -5 if v[2] > 5 else v[2]
        v[1:] += alpha * deltas[j] * .9
        
        
        
    
        
```
